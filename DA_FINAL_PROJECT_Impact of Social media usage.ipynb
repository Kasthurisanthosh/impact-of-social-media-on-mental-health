The Social Media Mental Health Indicators dataset contains anonymized data on social media usage patterns such as screen time, platform usage, and engagement levels. #Type of Analysis

**Exploratory Data Analysis (EDA)**

Used to understand the structure and distribution of the dataset by analyzing screen time, platform usage, and mental health indicators. Summary statistics and visualizations were created to identify trends, patterns, and outliers in the data.

**Correlation Analysis**

Applied to measure the strength and direction of relationships between social media usage variables (such as screen time and engagement) and mental health indicators like stress, anxiety, mood, and sleep quality.

**Comparative (Platform-wise) Analysis**

Performed to compare mental health indicators across different social media platforms and identify which platforms are associated with higher or lower stress, anxiety, or mood changes.

**Behavioral Segmentation Analysis**

Used to group users into distinct segments based on their social media behavior and mental health patterns, helping to identify high-risk and low-risk user groups.

**Predictive Analysis**

Implemented to predict mental health risk levels using social media usage behavior, enabling early identification of users potentially at higher risk.

**What features you used?**

mental_state – mental health status category

Number of Users – count of users in each category

**From this what you are showing us?**

The chart helps identify the prevalence of different mental health risk levels in the dataset, highlighting the proportion of users who may require further analysis or targeted interventions.

```python
# Bar Chart - Average Mental Health Indicators by Platform
platform_stress=df.groupby('platform')[['anxiety_level','stress_level','mood_level']].mean()
platform_stress.plot(kind='bar')
plt.xlabel('Platform')
plt.ylabel('Mental Health Indicator Level')
plt.title('Average Mental Health Indicators by Platform')
plt.show()
```

## Interpretation of the Chart: Average Mental Health Indicators by Platform (Bar Chart)

**Explain the Chart:**

This bar chart compares the average anxiety level, stress level, and mood level of users across different social media platforms. Each platform has separate bars representing the mean value of each mental health indicator.

**What is it saying?**

The chart shows that mental health indicators vary by platform. Some platforms are associated with higher average stress and anxiety levels, while others show comparatively better mood levels. This indicates that platform usage may influence users’ mental well-being differently.

**What features you used?**

platform – social media platform used

anxiety_level – average anxiety score

stress_level – average stress score

mood_level – average mood score

**From this what are we showing us?**

The chart highlights platform-wise differences in mental health impact, helping identify platforms that may require focused well-being interventions or further investigation.


```python
# HeatMap - Correlation Heatmap of Social Media Usage and Mental Health Indicators
numeric_cols = [
    'age',
    'daily_screen_time_hours',
    'social_media_time_hours',
    'negative_interactions_count',
    'positive_interactions_count',
    'sleep_hours',
    'physical_activity_min',
    'anxiety_level',
    'stress_level',
    'mood_level'
]

corr_data=df[numeric_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_data, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap of Social Media Usage and Mental Health Indicators')
plt.show()
```

## Interpretation of your above chart(HeatMap - Correlation Heatmap of Social Media Usage and Mental Health Indicators)

**Explain the Chart.**

This heatmap visualizes the correlation strength and direction between numerical variables related to social media usage, lifestyle factors, and mental health indicators. Color intensity and numeric values represent how strongly variables are related to each other.

**What is it saying?**

The chart shows that variables related to screen time and social media usage tend to have positive correlations with anxiety and stress levels, while healthy lifestyle factors such as sleep hours and physical activity show negative correlations with stress and anxiety.


**What features you used?**

Usage Metrics:
Daily screen time (hours), social media time (hours)

Interaction Metrics:
Positive and negative interaction counts

Lifestyle Factors:
Sleep hours, physical activity minutes

Mental Health Indicators:
Anxiety level, stress level, mood level

Demographic:
Age


**From this what you are showing us?**

Higher screen time and negative interactions are associated with higher anxiety and stress levels

Better sleep and physical activity are associated with lower stress and improved mood

Mood level is generally negatively correlated with anxiety and stress

This highlights how digital behavior and lifestyle habits are interconnected with mental well-being, helping identify potential risk and protective factors.



```python
# Pairplot of Social Media Usage and Mental Health Indicators
sns.pairplot(
    df,
    vars=[
        "daily_screen_time_min",
        "social_media_time_min",
        "stress_level",
        "anxiety_level",
        "sleep_hours",
        "physical_activity_min",
        "mood_level",
        "age",
        "negative_interactions_count",
        "positive_interactions_count"
    ],
    hue="mental_state")
plt.suptitle('Pairplot of Social Media Usage and Mental Health Indicators')
plt.show()
```

## Interpretation of your above chart(Pairplot of Social Media Usage and Mental Health Indicators)
**Explain the Chart**

This pairplot shows the relationships between multiple numerical variables related to social media usage, lifestyle factors, and mental health indicators. Each scatter plot compares two variables, while the diagonal plots show their distributions. Different colors represent different mental_state categories.

**What is it saying?**

The chart reveals patterns and relationships between variables such as screen time, stress, anxiety, sleep, and mood. It also shows how these relationships differ across mental health states, helping identify trends and separations between normal and high-risk users.

**What features were used?**

Social media usage: daily screen time, social media time

Mental health indicators: stress level, anxiety level, mood level

Lifestyle factors: sleep hours, physical activity

Interaction metrics: negative and positive interactions

Demographic: age

mental_state used as a color (hue) for comparison

**From this chart, what are we showing us?**

The pairplot helps identify correlations, clusters, and patterns across mental health states. It visually shows that users with higher screen time and negative interactions tend to have higher stress and anxiety, while better sleep and physical activity are associated with improved mood and lower risk.

```python
# Line Chart - Sleep Hours vs Average Stress Level
sleep_stress = df.groupby("sleep_hours")["stress_level"].mean().reset_index()
plt.plot(sleep_stress["sleep_hours"],sleep_stress["stress_level"],marker="o")
plt.title("Sleep Hours vs Average Stress Level")
plt.xlabel("Sleep Hours")
plt.ylabel("Average Stress Level")
plt.show()
```

## Interpretation of the above chart(Line Chart - Sleep Hours vs Average Stress Level)

**Explain the Chart:**

This line chart shows the relationship between sleep duration and the average stress level of users. The X-axis represents sleep hours, and the Y-axis shows the average stress level calculated for each sleep duration.

**What is it saying?**

The chart indicates that as sleep hours increase, the average stress level generally decreases. This suggests that users who get more sleep tend to experience lower stress.

**What features you used?**

sleep_hours – number of hours slept

stress_level – average stress level grouped by sleep hours

**From this what are we showing us?**

The chart highlights sleep as a protective factor, showing that adequate sleep is associated with reduced stress levels and better mental well-being.

```python
# Violin Plot - Stress Level Distribution by Platform (Grouped by Gender)
plt.figure(figsize=(10,6))
sns.violinplot(data=df,x="platform",y="stress_level",hue="gender")

plt.title("Stress Level Distribution by Platform (Grouped by Gender)")
plt.xlabel("Platform")
plt.ylabel("Stress Level")
plt.xticks(rotation=30)
plt.show()
```

## Interpretation of the Chart: Stress Level Distribution by Platform (Grouped by Gender – Violin Plot)

**Explain the Chart:**

This violin plot shows how stress levels are distributed across different social media platforms, further grouped by gender. The width of each violin represents the density of users at different stress levels, allowing comparison of both distribution and spread.

**What is it saying?**

The chart indicates that stress levels vary not only by platform but also by gender. For some platforms, one gender shows higher stress concentration or greater variability than the other, highlighting gender-based differences in stress experience.

**What features you used?**

platform – social media platform used

stress_level – users’ stress levels

gender – grouping variable to compare stress patterns by gender

**From this what are we showing us?**

The chart helps identify platforms where stress impact differs between genders, supporting more targeted analysis and gender-specific mental health insights.

```python
# KDE plot - positive interactions vs negative interactions
plt.figure(figsize=(10,6))

sns.kdeplot(df["positive_interactions_count"], fill=True, label="Positive Interactions")
sns.kdeplot(df["negative_interactions_count"], fill=True, label="Negative Interactions")

plt.title("Distribution of Positive vs Negative Interactions")
plt.xlabel("Interaction Count")
plt.ylabel("Density")
plt.legend()

plt.show()
```

## Interpretation of the Chart: Distribution of Positive vs Negative Interactions (KDE Plot)

**Explain the Chart:**

This KDE (density) plot compares the distribution of positive and negative interaction counts among users. It shows how frequently different interaction levels occur, using smooth curves instead of bars.

**What is it saying?**

The chart indicates that positive interactions generally occur more frequently than negative interactions. Negative interactions are more concentrated at lower counts, while positive interactions show a wider spread, suggesting users receive more positive engagement overall.

**What features you used?**

positive_interactions_count – number of positive interactions

negative_interactions_count – number of negative interactions

Density – distribution of interaction frequency

**From this what are we showing us?**

The chart highlights the balance between positive and negative social media experiences, helping assess how online interactions may influence users’ mental health.
